```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q='

 
import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()

```

### only for stations within 500m of Time Square!

```python
test_time_squre = '''

SELECT start_station_id, start_station_name,COUNT(cartodb_id) as number_of_trips, (ACOS( SIN(RADIANS(40.7577)) * SIN(RADIANS(citibike.start_station_latitude)) 
                       + COS(RADIANS(40.7577)) * COS(RADIANS(citibike.start_station_latitude)) * 
                       COS(RADIANS(citibike.start_station_longitude) - RADIANS(-73.9857)) ) * 6371000) AS distance_to_Time_Squre 
FROM citibike
WHERE (((ACOS( SIN(RADIANS(40.7577)) * SIN(RADIANS(citibike.start_station_latitude)) +
              COS(RADIANS(40.7577)) * COS(RADIANS(citibike.start_station_latitude)) * 
              COS(RADIANS(citibike.start_station_longitude) - RADIANS(-73.9857)) ) * 6371000) <= 500) OR 
                  (citibike.start_station_latitude = 40.7577 AND citibike.start_station_longitude = -73.9857))
GROUP BY start_station_id,start_station_name,start_station_latitude,start_station_longitude
'''


table_time_squre  = pd.read_csv(StringIO.StringIO(queryCartoDB(test_time_squre)), sep=',')
table_time_squre

#reference of converting radius to meters: 
# https://trustconverter.com/en/length-conversion/earth-radius/earth-radius-to-meters.html?value=1
# https://gist.github.com/robotmay/851250
```

## Task 4  Putting it all together

### Find the station that had the longest average trip duration during weekends and within 500m of TimeSquare!

```python
test_time_squre2 = '''

SELECT start_station_id, start_station_name, COUNT(cartodb_id)as number_of_trips, avg(tripduration) as weekends_avg_duration
FROM citibike
WHERE EXTRACT(DOW FROM starttime::date) in (0,6) AND
(((ACOS( SIN(RADIANS(40.7577)) * SIN(RADIANS(citibike.start_station_latitude)) +
              COS(RADIANS(40.7577)) * COS(RADIANS(citibike.start_station_latitude)) * 
              COS(RADIANS(citibike.start_station_longitude) - RADIANS(-73.9857)) ) * 6371000) <= 500) OR 
                  (citibike.start_station_latitude = 40.7577 AND citibike.start_station_longitude = -73.9857))

GROUP BY start_station_id,start_station_name,start_station_latitude,start_station_longitude
ORDER BY weekends_avg_duration DESC
Limit 1
'''
table_time_squre2  = pd.read_csv(StringIO.StringIO(queryCartoDB(test_time_squre2)), sep=',')
table_time_squre2
```
